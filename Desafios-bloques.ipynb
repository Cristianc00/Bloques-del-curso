### Desafio 2

```python
def contar_pares(lista_de_numeros):
    """
    Cuenta la cantidad de números pares en una lista de enteros.
    
    Args:
        lista_de_numeros (list): Una lista que contiene números enteros.
        
    Returns:
        int: El número total de elementos pares encontrados.
    """
    # Inicializa el contador de números pares en cero.
    contador_pares = 0
    
    # Itera sobre cada número en la lista.
    for numero in lista_de_numeros:
        # La condición de paridad: un número es par si el residuo
        # de la división por 2 es 0. El operador '%' es el módulo.
        if numero % 2 == 0:
            # Si es par, incrementa el contador.
            contador_pares += 1
            
    # Devuelve el resultado final.
    return contador_pares

# --- Ejemplos de Uso ---

# Lista con números pares e impares
numeros_1 = [1, 4, 7, 2, 9, 6, 8, 10]
resultado_1 = contar_pares(numeros_1)

print(f"La lista es: {numeros_1}")
# Esperado: 4, 2, 6, 8, 10 (5 números)
print(f"El número de elementos pares es: {resultado_1}")

# Lista vacía
numeros_2 = []
resultado_2 = contar_pares(numeros_2)

print(f"\nLa lista es: {numeros_2}")
# Esperado: 0
print(f"El número de elementos pares es: {resultado_2}")
```

```python
def clasificar_tres(a, b, c):
    """Clasifica tres números distintos en orden ascendente."""
    # Comprueba la posición del número más pequeño (a, b o c)
    
    if a < b and b < c:
        return (a, b, c)  # Caso 1: a < b < c
    elif a < c and c < b:
        return (a, c, b)  # Caso 2: a < c < b
        
    elif b < a and a < c:
        return (b, a, c)  # Caso 3: b < a < c
    elif b < c and c < a:
        return (b, c, a)  # Caso 4: b < c < a
        
    elif c < a and a < b:
        return (c, a, b)  # Caso 5: c < a < b
    else:
        return (c, b, a)  # Caso 6: c < b < a (el último caso posible)

# --- Ejemplo de Uso ---
num1, num2, num3 = 70, 20, 50
ordenado = clasificar_tres(num1, num2, num3)

print(f"Números: ({num1}, {num2}, {num3})")
print(f"Orden Ascendente: {ordenado}")
```

### Desafio 2 (4)

```python
def calcular_promedio_5_calificaciones():
    """
    Pide al usuario 5 calificaciones y calcula su promedio.
    """
    calificaciones = []
    
    # 1. Leer las 5 calificaciones
    print("Ingresa 5 calificaciones para calcular el promedio:")
    for i in range(5):
        while True:
            try:
                # Pedir la calificación al usuario
                calif = float(input(f"Calificación #{i + 1}: "))
                calificaciones.append(calif)
                break  # Salir del bucle while si la entrada es válida
            except ValueError:
                # Manejar el error si el usuario no ingresa un número
                print("⚠️ Error: Por favor, ingresa un número válido.")
                
    # 2. Calcular la suma y el promedio
    # Usamos las funciones sum() y len() que Python trae integradas
    suma_calificaciones = sum(calificaciones)
    cantidad_calificaciones = len(calificaciones) # Siempre será 5
    
    # Prevenir división por cero (aunque la lista siempre tendrá 5 elementos aquí)
    if cantidad_calificaciones > 0:
        promedio = suma_calificaciones / cantidad_calificaciones
        # 3. Mostrar el resultado
        print("-" * 30)
        print(f"Suma total: {suma_calificaciones:.2f}")
        print(f"Cantidad de calificaciones: {cantidad_calificaciones}")
        print(f"El **promedio** de las 5 calificaciones es: **{promedio:.2f}**")
        print("-" * 30)
        return promedio
    else:
        print("No se ingresaron calificaciones.")
        return 0

# Ejecutar la función
calcular_promedio_5_calificaciones()
```

```python
def verificar_multiplos(numero_dado):
    """
    Verifica si un número dado es múltiplo de 2, 3, 5, 7, 9, 10, y 11.
    """
    
    # Lista de divisores a verificar
    divisores = [2, 3, 5, 7, 9, 10, 11]
    
    print(f"--- Verificación de Múltiplos para el número {numero_dado} ---")

    # Iterar sobre cada divisor y realizar la verificación
    for divisor in divisores:
        # La condición es: si el residuo de la división es 0, es múltiplo.
        es_multiplo = numero_dado % divisor == 0
        
        # Determinar la respuesta para la impresión
        respuesta = "Sí" if es_multiplo else "No"
        
        # Imprimir el resultado
        print(f"¿Es {numero_dado} múltiplo de {divisor}? {respuesta}")

# --- Ejecución del Desafío con el Ejemplo ---
numero_ejemplo = 30
verificar_multiplos(numero_ejemplo)

print("\n--- Otra Prueba (ej. 154) ---")
verificar_multiplos(154)
```

### Desafio 3 (10)

```python
def generar_factura():
    productos = []
    while True:
        nombre_producto = input("Ingrese el nombre del producto (o 'fin' para terminar): ").strip()
        if nombre_producto.lower() == 'fin':
            break

        try:
            cantidad = int(input(f"Ingrese la cantidad de '{nombre_producto}': "))
            precio = float(input(f"Ingrese el precio unitario de '{nombre_producto}': "))
            productos.append({'nombre': nombre_producto, 'cantidad': cantidad, 'precio': precio})
        except ValueError:
            print("¡Error! Por favor, ingrese una cantidad y un precio válidos.")

    if not productos:
        print("\nNo se agregaron productos a la factura.")
        return

    print("\n" + "=" * 40)
    print("{:^40}".format("FACTURA"))
    print("-" * 40)
    print("{:<20} {:>8} {:>10}".format("Producto", "Cant.", "Precio"))
    print("-" * 40)

    total_factura = 0
    for producto in productos:
        nombre = producto['nombre']
        cantidad = producto['cantidad']
        precio = producto['precio']
        subtotal = cantidad * precio
        print("{:<20} {:>8} {:>10.2f}".format(nombre, cantidad, precio))
        total_factura += subtotal

    print("-" * 40)
    print("{:>30} {:>10.2f}".format("Total:", total_factura))
    print("=" * 40)
    print("{:^40}".format("¡Gracias por su compra!"))

if __name__ == "__main__":
    generar_factura()
```

```python
import os

# --- 1. Preparación: Crear un archivo de ejemplo (INPUT) ---
# Simulamos un archivo de calificaciones (una por línea)
archivo_entrada = "calificaciones.txt"

datos_ejemplo = "8.5\n9.0\n7.5\n10.0\n6.0"
with open(archivo_entrada, 'w') as f:
    f.write(datos_ejemplo)

print(f"✅ Archivo de entrada '{archivo_entrada}' creado con datos de ejemplo.")

# --- 2. Procesamiento: Leer, Calcular Promedio ---

def calcular_y_guardar_promedio(archivo_in, archivo_out):
    """Lee calificaciones, calcula el promedio y lo escribe en un nuevo archivo."""
    
    suma_calificaciones = 0.0
    conteo_calificaciones = 0
    
    # Lectura del archivo de entrada (calificaciones.txt)
    try:
        with open(archivo_in, 'r') as f_in:
            for linea in f_in:
                try:
                    # Convertir la línea (string) a número flotante
                    calificacion = float(linea.strip())
                    suma_calificaciones += calificacion
                    conteo_calificaciones += 1
                except ValueError:
                    # Opcional: Manejar líneas vacías o no numéricas
                    print(f"⚠️ Advertencia: Omitiendo línea no válida: {linea.strip()}")
                    
    except FileNotFoundError:
        print(f"❌ Error: El archivo de entrada '{archivo_in}' no fue encontrado.")
        return

    # Cálculo del promedio
    if conteo_calificaciones > 0:
        promedio = suma_calificaciones / conteo_calificaciones
    else:
        promedio = 0.0
        
    # Escritura del archivo de salida (promedio_final.txt)
    with open(archivo_out, 'w') as f_out:
        f_out.write(f"El promedio de las calificaciones es: {promedio:.2f}")

    print(f"\n✨ Cálculo completado.")
    print(f"El promedio calculado es: {promedio:.2f}")
    print(f"El resultado ha sido guardado en '{archivo_out}'.")

# --- 3. Ejecución ---
archivo_salida = "promedio_final.txt"
calcular_y_guardar_promedio(archivo_entrada, archivo_salida)

# Opcional: Limpiar archivos de prueba si es necesario (para mantener el directorio ordenado)
# os.remove(archivo_entrada)
# os.remove(archivo_salida)
```

### Problema 5

```python
import json

# --- 1. Preparación: Crear un archivo JSON de ejemplo (INPUT) ---
archivo_entrada = "calificaciones.json"

datos_ejemplo = {
    "alumnos": [
        {"nombre": "Ana", "calificaciones": [8.5, 9.0, 7.5]},
        {"nombre": "Luis", "calificaciones": [6.0, 8.0, 7.0]},
        {"nombre": "Maria", "calificaciones": [10.0, 9.5, 10.0]}
    ]
}

# Escribir el diccionario de Python al archivo JSON
with open(archivo_entrada, 'w') as f:
    json.dump(datos_ejemplo, f, indent=4)

print(f"✅ Archivo de entrada '{archivo_entrada}' creado con datos de ejemplo.")

# --- 2. Procesamiento: Leer, Calcular Promedio ---

def calcular_y_guardar_promedio_json(archivo_in, archivo_out):
    """Lee un JSON, calcula el promedio general y lo escribe en un nuevo JSON."""
    
    todas_las_calificaciones = []
    
    # 2a. Leer el archivo JSON
    try:
        with open(archivo_in, 'r') as f_in:
            data = json.load(f_in)
            
            # 2b. Extraer todas las calificaciones
            for alumno in data.get('alumnos', []):
                # Usar extend para añadir todas las calificaciones de la lista a la lista general
                todas_las_calificaciones.extend(alumno.get('calificaciones', []))
                
    except FileNotFoundError:
        print(f"❌ Error: El archivo de entrada '{archivo_in}' no fue encontrado.")
        return
    except json.JSONDecodeError:
        print(f"❌ Error: El archivo '{archivo_in}' no es un JSON válido.")
        return

    # 2c. Calcular el promedio general
    total_calificaciones = len(todas_las_calificaciones)
    if total_calificaciones > 0:
        suma = sum(todas_las_calificaciones)
        promedio = suma / total_calificaciones
    else:
        promedio = 0.0
        
    # 3. Escritura del archivo JSON de salida
    resultado = {
        "promedio_general_clase": round(promedio, 2),
        "total_calificaciones_procesadas": total_calificaciones
    }
    
    with open(archivo_out, 'w') as f_out:
        # Usar json.dump para escribir el diccionario de resultados en formato JSON
        json.dump(resultado, f_out, indent=4)

    print(f"\n✨ Cálculo completado.")
    print(f"El promedio general es: {promedio:.2f}")
    print(f"El resultado ha sido guardado en '{archivo_out}'.")

# --- 4. Ejecución ---
archivo_salida = "promedio_final.json"
calcular_y_guardar_promedio_json(archivo_entrada, archivo_salida)
```

```python
import random

def simular_carrera(nombres_autos, duracion=10):
  velocidades = {nombre: random.randint(5, 15) for nombre in nombres_autos} # Velocidad entre 5 y 15 m/s
  distancias = {nombre: 0 for nombre in nombres_autos}

  print("¡Comienza la carrera!")
  for segundo in range(1, duracion + 1):
    print(f"\n--- Segundo {segundo} ---")
    for nombre in nombres_autos:
      velocidad = velocidades[nombre]
      distancias[nombre] += velocidad
      print(f"{nombre} avanzó {velocidad} metros. Distancia total: {distancias[nombre]} metros.")

  # Determinar el ganador o los ganadores
  distancia_maxima = max(distancias.values())
  ganadores = [nombre for nombre, distancia in distancias.items() if distancia == distancia_maxima]

  print("\n--- ¡Carrera Terminada! ---")
  if len(ganadores) == 1:
    print(f"¡El ganador es {ganadores[0]} con una distancia de {distancia_maxima} metros!")
  else:
    print(f"¡Empate entre {', '.join(ganadores)} con una distancia de {distancia_maxima} metros!")

  return distancias

# Nombres de los autos participantes
autos = ["Rayo McQueen", "Chick Hicks", "Strip Weathers", "Meteoro"]

# Simular la carrera
resultados = simular_carrera(autos)
```

```python
def calcular_mcd(a, b):
    """
    Calcula el Máximo Común Divisor (MCD) de dos números enteros (a y b) 
    utilizando el Algoritmo de Euclides (recursivo).
    """
    # Condición base: Si b es 0, el MCD es a.
    if b == 0:
        return a
    else:
        # Llamada recursiva: MCD(a, b) = MCD(b, a % b)
        return calcular_mcd(b, a % b)

# --- Ejemplos de Uso ---
num1 = 54  # 54 = 2 * 3^3
num2 = 24  # 24 = 2^3 * 3
# El MCD es 2 * 3 = 6

mcd_resultado_1 = calcular_mcd(num1, num2)
print(f"El MCD de {num1} y {num2} es: {mcd_resultado_1}") # Esperado: 6

num3 = 105
num4 = 30
# El MCD es 15

mcd_resultado_2 = calcular_mcd(num3, num4)
print(f"El MCD de {num3} y {num4} es: {mcd_resultado_2}") # Esperado: 15
```

```python
import collections
import string

def analizar_y_mostrar_top_palabras(texto):
    """
    Analiza un texto para encontrar las 10 palabras más comunes y muestra su frecuencia.
    """
    # 1. Preparación del texto: convertir a minúsculas y eliminar puntuación
    # str.maketrans se usa para crear una tabla de traducción que mapea
    # cada signo de puntuación a None (eliminación)
    traductor = str.maketrans('', '', string.punctuation)
    texto_limpio = texto.lower().translate(traductor)
    
    # 2. Obtener una lista de palabras (excluyendo cadenas vacías si las hay)
    palabras = [palabra for palabra in texto_limpio.split() if palabra]
    
    # 3. Contar la frecuencia de cada palabra usando Counter
    conteo_palabras = collections.Counter(palabras)
    
    # 4. Obtener las 10 palabras más comunes
    top_10 = conteo_palabras.most_common(10)
    
    print("--- Top 10 Palabras Más Comunes y Frecuencia ---")
    
    if not top_10:
        print("El texto está vacío o no contiene palabras válidas.")
        return

    # Mostrar los resultados
    for palabra, frecuencia in top_10:
        print(f"'{palabra}': {frecuencia} veces")

# --- Texto de Ejemplo ---
texto_analisis = """
Python es un lenguaje de programación muy popular. Python se utiliza para 
ciencia de datos, desarrollo web y automatización. La popularidad de Python 
sigue creciendo, haciendo que Python sea una excelente opción para nuevos 
programadores. Este texto corto usa la palabra Python varias veces.
"""

analizar_y_mostrar_top_palabras(texto_analisis)
```

```python
from datetime import date
from typing import List, Optional

class Autor:
    def __init__(self, nombre: str, fecha_nacimiento: str, nacionalidad: str = "Desconocida"):
        self.nombre = nombre
        # Almacenamos la fecha como objeto date para cálculos precisos (AAAA-MM-DD)
        try:
            año, mes, dia = map(int, fecha_nacimiento.split('-'))
            self.fecha_nacimiento = date(año, mes, dia)
        except:
            # En caso de error de formato, usa una fecha por defecto o levanta una excepción
            print("Advertencia: Formato de fecha incorrecto. Usando 1900-01-01.")
            self.fecha_nacimiento = date(1900, 1, 1)

        self.libros_escritos: List[str] = []
        self.nacionalidad: str = nacionalidad
        self.generos_principales: List[str] = [] # Atributo adicional
        self.es_activo: bool = True               # Atributo adicional

    def agregar_genero(self, genero: str):
        """Método adicional para agregar un género principal."""
        if genero not in self.generos_principales:
            self.generos_principales.append(genero)

    def calcular_edad(self) -> int:
        """
        Método adicional para calcular la edad exacta en años.
        Usa la fecha actual del sistema.
        """
        hoy = date.today()
        # Restamos los años y ajustamos si el cumpleaños aún no ha pasado este año
        edad = hoy.year - self.fecha_nacimiento.year
        if (hoy.month, hoy.day) < (self.fecha_nacimiento.month, self.fecha_nacimiento.day):
            edad -= 1
        return edad

    def mostrar_info_resumida(self):
        """Método adicional para imprimir información clave."""
        edad = self.calcular_edad()
        estado = "Activo" if self.es_activo else "Inactivo"
        print("--- Información del Autor ---")
        print(f"Nombre: **{self.nombre}** ({self.nacionalidad})")
        print(f"Edad: {edad} años (Nacimiento: {self.fecha_nacimiento.strftime('%Y-%m-%d')})")
        print(f"Géneros: {', '.join(self.generos_principales)}")
        print(f"Estado: {estado}")

# Uso y Ejemplo Corregido
autor_gabo = Autor("Gabriel García Márquez", "1927-03-06", "Colombiano")
autor_gabo.agregar_genero("Realismo Mágico")
autor_gabo.agregar_genero("Ficción")
autor_gabo.es_activo = False # Para un autor fallecido

autor_gabo.mostrar_info_resumida()
```

```python
import math

def factorial_con_excepciones(n):
    """
    Calcula el factorial de un número entero positivo.
    Maneja excepciones para números negativos, no enteros, y números grandes.

    Args:
        n (any): El número para el cual se calculará el factorial.

    Returns:
        int or None: El factorial del número si es válido, de lo contrario None.
    """
    try:
        # 1. Validar que el número es un entero
        if not isinstance(n, int):
            raise TypeError("El valor ingresado debe ser un número entero.")
        
        # 2. Validar que el número no es negativo
        if n < 0:
            raise ValueError("El factorial no está definido para números negativos.")
        
        # 3. Validar si el número es demasiado grande para evitar desbordamiento
        # El límite para el tipo de dato int en Python es flexible, pero esta
        # validación es útil para evitar cálculos excesivamente largos.
        # Se establece un límite práctico, por ejemplo, 1000.
        if n > 1000:
            raise OverflowError("El número es demasiado grande para ser procesado eficientemente.")
        
        # 4. Calcular el factorial
        # math.factorial es una forma eficiente de calcularlo
        return math.factorial(n)

    except TypeError as e:
        print(f"Error de Tipo: {e}")
        return None
    except ValueError as e:
        print(f"Error de Valor: {e}")
        return None
    except OverflowError as e:
        print(f"Error de Desbordamiento: {e}")
        return None
    except Exception as e:
        # Captura cualquier otra excepción inesperada
        print(f"Ha ocurrido un error inesperado: {e}")
        return None

# --- Ejemplos de uso ---
print("Ejemplo de uso:")
print(f"Factorial de 5: {factorial_con_excepciones(5)}")
print(f"Factorial de 0: {factorial_con_excepciones(0)}")
print("-" * 20)

print("Ejemplos de errores:")
print(f"Factorial de -3: {factorial_con_excepciones(-3)}")
print(f"Factorial de 4.5: {factorial_con_excepciones(4.5)}")
print(f"Factorial de 'abc': {factorial_con_excepciones('abc')}")
print(f"Factorial de 1500: {factorial_con_excepciones(1500)}")
```

```python
import math

# 1. Creación de la Excepción Personalizada
class NegativeNumberError(Exception):
    """Excepción disparada al intentar calcular la raíz de un número negativo."""
    def __init__(self, numero):
        self.message = f"Error: No se puede calcular la raíz cuadrada de un número negativo ({numero})."
        super().__init__(self.message)

def calcular_raiz(numero):
    if numero < 0:
        # 2. Disparo de la Excepción
        raise NegativeNumberError(numero)
    return math.sqrt(numero)

# 3. Implementación del Manejo
try:
    num_negativo = -4
    resultado = calcular_raiz(num_negativo)
    print(f"Raíz cuadrada: {resultado}")
except NegativeNumberError as e:
    # Captura la excepción personalizada
    print(e)
except TypeError:
    # Manejo de otros posibles errores (ej. si se pasa un texto)
    print("Error: El valor debe ser un número.")

# Ejecución de prueba con un número positivo
try:
    print(f"\nRaíz cuadrada de 9: {calcular_raiz(9)}")
except NegativeNumberError as e:
    pass # No debería dispararse
```

```python
class Nodo:
    def __init__(self, valor):
        self.valor = valor
        self.izq = None
        self.der = None

def preorden(nodo):
    """Recorrido en Preorden: Raíz -> Izquierda -> Derecha."""
    if nodo:
        print(nodo.valor, end=" ") # 1. Raíz
        preorden(nodo.izq)          # 2. Izquierda
        preorden(nodo.der)          # 3. Derecha

# Construcción del Árbol (3 niveles)
raiz = Nodo(10)
raiz.izq = Nodo(5)
raiz.der = Nodo(15)
raiz.izq.izq = Nodo(3)
raiz.izq.der = Nodo(7)
raiz.der.der = Nodo(20)

print("Recorrido en Preorden:")
preorden(raiz)
print()
```

```python
# Usamos la clase Nodo del Desafío 76

def encontrar_max_postorden(nodo):
    """Recorre en Postorden y devuelve el valor máximo."""
    if not nodo:
        # Usamos -infinito para que cualquier valor del nodo sea mayor
        return float('-inf')

    # 1. Izquierda (llamada recursiva)
    max_izq = encontrar_max_postorden(nodo.izq)
    # 2. Derecha (llamada recursiva)
    max_der = encontrar_max_postorden(nodo.der)
    # 3. Raíz (comparación)
    
    # El máximo es el mayor entre el nodo actual, el máximo izquierdo y el máximo derecho
    return max(nodo.valor, max_izq, max_der)

# Usando el árbol del Desafío 76: el máximo es 20
valor_maximo = encontrar_max_postorden(raiz)
print(f"Valor máximo encontrado (Postorden): {valor_maximo}")
```

```python
class NodoExpresion:
    def __init__(self, valor):
        self.valor = valor # Operador o número
        self.izq = None
        self.der = None

def evaluar_arbol(nodo):
    """Evalúa el árbol de expresiones de forma recursiva."""
    # Si es un número (hoja), devuelve el valor
    if isinstance(nodo.valor, int):
        return nodo.valor

    # Si es un operador, evalúa los subárboles
    operador = nodo.valor
    val_izq = evaluar_arbol(nodo.izq)
    val_der = evaluar_arbol(nodo.der)

    if operador == '+':
        return val_izq + val_der
    elif operador == '-':
        return val_izq - val_der
    elif operador == '*':
        return val_izq * val_der
    elif operador == '/':
        # Manejo simple de división (puedes añadir manejo de ZeroDivisionError)
        return val_izq / val_der
    
    return 0

# Construcción del Árbol para "5 + 3 * 4" (asumiendo orden de precedencia)
# El nodo principal es '+'
raiz_exp = NodoExpresion('+')
raiz_exp.izq = NodoExpresion(5) # El 5 está a la izquierda
raiz_exp.der = NodoExpresion('*') # El '*' está a la derecha
raiz_exp.der.izq = NodoExpresion(3)
raiz_exp.der.der = NodoExpresion(4)

# Evaluación
resultado = evaluar_arbol(raiz_exp)

print(f"Resultado de la expresión '5 + 3 * 4': {resultado}")
```

```python
from collections import deque

class Grado:
    def __init__(self, nombre_grado, estudiantes):
        self.nombre = nombre_grado
        self.estudiantes = estudiantes
        self.hijos = [] # Grados inferiores (e.g., Grado 11)

def recorrido_por_niveles(raiz):
    """Implementa el recorrido BFS (por niveles)."""
    if not raiz:
        return

    cola = deque([raiz])
    
    while cola:
        nodo_actual = cola.popleft()
        
        print(f"Grado: **{nodo_actual.nombre}** -> Estudiantes: {nodo_actual.estudiantes}")
        
        # Añadir los hijos (niveles inferiores) a la cola
        for hijo in nodo_actual.hijos:
            cola.append(hijo)

# Construcción del árbol (simulación de jerarquía)
g12 = Grado("Grado 12", ["Alice", "Bob"])
g11_a = Grado("Grado 11A", ["Charlie"])
g11_b = Grado("Grado 11B", ["Dana", "Eve"])
g10 = Grado("Grado 10", ["Frank"])

g12.hijos.extend([g11_a, g11_b])
g11_a.hijos.append(g10)

recorrido_por_niveles(g12)
```

```python
def busqueda_binaria_estudiante(lista_nombres, nombre_buscado):
    """Implementa búsqueda binaria en la lista ordenada."""
    bajo = 0
    alto = len(lista_nombres) - 1
    
    while bajo <= alto:
        medio = (bajo + alto) // 2
        
        if lista_nombres[medio] == nombre_buscado:
            return f"Estudiante **{nombre_buscado}** encontrado en la posición **{medio}**."
        elif lista_nombres[medio] < nombre_buscado:
            bajo = medio + 1  # Busca en la mitad derecha
        else:
            alto = medio - 1  # Busca en la mitad izquierda
            
    return f"Estudiante **{nombre_buscado}** no encontrado."

estudiantes_ordenados = ["Ana", "Carlos", "David", "Elena", "Sofía"]

print(busqueda_binaria_estudiante(estudiantes_ordenados, "David"))
print(busqueda_binaria_estudiante(estudiantes_ordenados, "Javier"))
```

```python
def ordenar_por_seleccion_promedio(estudiantes):
    """Ordena la lista de estudiantes por promedio (descendente) usando Selección."""
    n = len(estudiantes)
    
    for i in range(n):
        # Encontramos el índice del elemento más alto en la sublista no ordenada [i:n]
        indice_max = i 
        for j in range(i + 1, n):
            # Comparamos promedios (el segundo elemento de la tupla)
            if estudiantes[j][1] > estudiantes[indice_max][1]:
                indice_max = j
        
        # Intercambiamos el elemento más alto con el elemento en la posición actual i
        estudiantes[i], estudiantes[indice_max] = estudiantes[indice_max], estudiantes[i]
    
    return estudiantes

lista_estudiantes = [("Luis", 8.5), ("Maria", 9.2), ("Juan", 7.8), ("Ana", 9.5)]
ordenados = ordenar_por_seleccion_promedio(lista_estudiantes)

print("Estudiantes ordenados por promedio (Descendente):")
for nombre, promedio in ordenados:
    print(f"- {nombre}: {promedio}")
```

```python
from collections import Counter
import re # Para limpieza de palabras

def contar_palabras_comunes(nombre_archivo, num_palabras=5):
    """Lee un archivo y muestra las 'num_palabras' más comunes."""
    try:
        with open(nombre_archivo, "r", encoding="utf-8") as f:
            texto = f.read().lower()
            
            # Limpieza y tokenización: reemplaza lo que no sea letra/espacio por espacio
            texto_limpio = re.sub(r'[^\w\s]', '', texto)
            
            # Divide el texto en palabras y filtra las vacías
            palabras = [palabra for palabra in texto_limpio.split() if palabra]
            
            # 1. Contar frecuencias con Counter
            contador = Counter(palabras)
            
            # 2. Obtener las N palabras más comunes
            mas_comunes = contador.most_common(num_palabras)
            
            print(f"\n--- {num_palabras} Palabras más comunes ---")
            for palabra, cuenta in mas_comunes:
                print(f"**{palabra}**: {cuenta} veces")
                
    except FileNotFoundError:
        print(f"Error: El archivo '{nombre_archivo}' no fue encontrado.")

# Ejemplo de uso
contar_palabras_comunes("libros.txt")
```
